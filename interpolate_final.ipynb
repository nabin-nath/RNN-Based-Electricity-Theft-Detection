```python
import pandas as pd
import numpy as np

df = pd.read_csv('/content/drive/MyDrive/Major Project - II/1.csv')
df = df.replace(0, np.nan)
```

```python
df.head()
```

```python
df2 = pd.DataFrame(df.T)
df2.reset_index(inplace=True)
df3 = df2.iloc[2:, :]
df3['index'] = pd.to_datetime(df3['index'])
df3.head()
```

```python
df3.set_index('index', inplace=True)
df3.head()
```

```python
temp = df3.astype('float64')
```

```python
temp.sort_index(ascending=True, inplace=True)
```

```python
temp.interpolate(method='pchip', inplace=True)
```

```python
temp.reset_index(inplace=True)
```

```python
temp.head()
```

```python
temp = temp.astype('object')
```

```python
final_df = pd.concat([df2.iloc[:2, :], temp])
final_df.head()
```

```python
final_df = final_df.T.fillna(0)
final_df.head()
```

```python
final_df.to_csv("/content/drive/MyDrive/Major Project - II/1_interpolate_n.csv")
```
